## Option C: Practice Datasets (Start Here!)
### No download needed — runs immediately to test your environment

```python
import scanpy as sc
import squidpy as sq
import matplotlib.pyplot as plt

print('Testing your Python environment...')
print(f'  Scanpy version:  {sc.__version__}')
print(f'  Squidpy version: {sq.__version__}')
```

```python
# ── Practice Dataset 1: Human PBMCs (great for learning scRNA-seq clustering) ──
# This downloads ~10 MB — a subset of 3000 human blood immune cells
print('Downloading PBMC3k practice dataset...')
pbmc = sc.datasets.pbmc3k()

print(f'\n✅ PBMC dataset loaded!')
print(f'   Cells: {pbmc.n_obs:,}')
print(f'   Genes: {pbmc.n_vars:,}')
print('\nThis dataset has B cells, T cells, NK cells, Monocytes — perfect')
print('for practicing the Module 1 workflow before switching to tumor data.')
```

```python
# ── Practice Dataset 2: Visium Mouse Brain (built-in Squidpy dataset) ──
# Not tumor tissue, but great for learning the spatial workflow
print('Downloading Visium mouse brain practice dataset...')
visium_brain = sq.datasets.visium_hne_adata()

print(f'\n✅ Visium brain dataset loaded!')
print(f'   Spots: {visium_brain.n_obs:,}')
print(f'   Genes: {visium_brain.n_vars:,}')
print(f'   Has spatial coords: {"spatial" in visium_brain.obsm}')

# Quick sanity check — show the H&E image with spots
sc.pl.spatial(
    visium_brain,
    img_key='hires',
    color=None,
    title='Practice Visium Data: Mouse Brain (H&E + spots)',
    save='_00_practice_visium_check.png'
)
print('\nIf you see a brain tissue image above, your spatial environment is working!')
```

```python
# Save practice datasets for use in Modules 1 and 2
pbmc.write_h5ad('data/raw/scrna/practice_pbmc3k.h5ad')
visium_brain.write_h5ad('data/raw/visium/practice_visium_brain.h5ad')

print('✅ Practice datasets saved!')
print('   → Partner A: open data/raw/scrna/practice_pbmc3k.h5ad in Notebook 01')
print('   → Partner B: open data/raw/visium/practice_visium_brain.h5ad in Notebook 02')
```

## Option A: Real Tumor Data — Annotated Kidney & Lung Cancer with TLS
### (Download this when you're ready for the real project)

**Source:** Zenodo record 14620362 — the same dataset used by the BostonGene reference project

**Contains:**
- 3 Kidney cancer + 5 Lung cancer 10x Visium samples
- Expert pathologist annotations of TLS regions (ground truth!)
- H&E images

**Manual download steps:**
1. Go to: https://zenodo.org/records/14620362
2. Click the green **Download** button or download individual files
3. Extract/move files to `data/raw/visium/`

**Or try the automated download below:**

```python
# Automated Zenodo download
# Note: This may take 10-30 min depending on your internet speed (~4GB total)

import urllib.request
import zipfile

ZENODO_URL = 'https://zenodo.org/records/14620362/files/'
SAVE_DIR   = Path('data/raw/visium')

# List of files on the Zenodo record (update if Zenodo page changes)
# Check https://zenodo.org/records/14620362 for the exact filenames
FILES_TO_DOWNLOAD = [
    # Add filenames from the Zenodo page here, e.g.:
    # 'kidney_sample1.tar.gz',
    # 'lung_sample1.tar.gz',
]

if not FILES_TO_DOWNLOAD:
    print('⚠️  Please visit https://zenodo.org/records/14620362')
    print('    and add the exact filenames to FILES_TO_DOWNLOAD above.')
    print('    (Or download manually from the webpage and put in data/raw/visium/)')
else:
    for filename in FILES_TO_DOWNLOAD:
        url  = ZENODO_URL + filename
        dest = SAVE_DIR / filename
        if dest.exists():
            print(f'  Already downloaded: {filename}')
            continue
        print(f'  Downloading {filename}...')
        urllib.request.urlretrieve(url, dest)
        print(f'  ✅ Done: {filename}')
```
